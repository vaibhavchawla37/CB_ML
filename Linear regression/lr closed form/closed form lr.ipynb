```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)
```

```python
Y=Y.reshape(-1,1)

print(X.shape)
print(Y.shape)
```

```python
X=(X-X.mean())/X.std()
```

```python
plt.figure()
plt.title('Normalized Data')
plt.style.use('seaborn')
plt.scatter(X,Y)
```

```python
ones=np.ones((X.shape[0],1))
X_=np.hstack((X,ones))
np.dot(X_.T,X_)
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def get_theta_cf(X,Y):
    first_part = np.dot(X.T,X)
    second_part= np.dot(X.T,Y)
    theta = np.dot(np.linalg.pinv(first_part),second_part)
    return theta
```

```python
theta=get_theta_cf(X_,Y)
theta
```

```python
y_=predict(X_,theta)
plt.figure()
plt.title('Normalized Data')
plt.style.use('seaborn')
plt.scatter(X,Y)
plt.plot(X,y_,color='orange')
```
