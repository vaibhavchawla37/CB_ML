```python
from urllib.request import urlopen
import numpy as np 
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films'
```

```python
print(url)
```

```python
data=urlopen(url)
data
```

```python
html=data.read()
print(html)
data.close()
```

```python
from bs4 import BeautifulSoup as soup
aveng_soup=soup(html,'html.parser')
aveng_soup
```

```python
print(type(aveng_soup))
```

```python
aveng_soup.h1
```

```python
aveng_soup.find_all('h1',{})
```

```python
tables=aveng_soup.find_all('table',{'class':'wikitable'})
aveng_table=tables[0]
len(tables)
```

```python
type(aveng_table)
```

```python
headers=aveng_table.find_all('th',{'scope':'col'})
len(headers)
print(type(headers))
```

```python
print(headers[0].text)
```

```python
cloumn_titles=[ct.text[:-1] for ct in headers[0:]]
print(cloumn_titles)
```

```python
rows_data=aveng_table.find_all('tr')[1:]
print(len(rows_data))
print(type(rows_data))
print(type(rows_data[0]))
```

```python
first_row=rows_data[1].find_all('td')
len(first_row)
for d in first_row:
    print(d.text[:- 1])
```

```python
table_rows=[]
for row in rows_data:
    row_data=row.find_all('td',{})
    curr_row=[]
    if (len(row_data)!=0):
        for idx,rdata in enumerate(row_data):
            curr_row.append(np.array(rdata.text[:-1]))
        table_rows.append(curr_row)
```

```python
row_data=rows_data[0].find_all('td',{})
print(len(row_data))
curr_row=[]
for idx,rdata in enumerate(row_data):
    if idx==0:
        curr_row.append(rdata.text[:-1])
```

```python
print(table_rows)
```

```python
len(table_rows[3])
```

```python
rows_data=aveng_table.find_all('tr')[1:]
trd=[]
for r in rows_data:
    rr=r.find_all('th',{'scope':'row'})
    cu=[]
    if (len(rr)!=0):
        for idx,cd in enumerate(rr):
            cu.append(cd.text[:-1])
        trd.append(cu)
```

```python
print(len(trd))
```

```python
np.array(trd)
```

```python
print(table_rows)
```

```python
np.array(trd)
```

```python
x=pd.DataFrame(trd)
y=pd.DataFrame(table_rows)
y=pd.concat([x,y],axis=1)

y.set_axis(cloumn_titles, axis=1, inplace=True)
y.to_csv('test.csv',index=False)
```
